```python
#Input data 
ess = pd.read_csv('/Users/li.13503/Desktop/heatmap/CRISPRGeneEffect.csv', index_col = 0).sort_index(axis = 0) # 17386 genes
shangjia = pd.read_csv('/Users/li.13503/Downloads/pathway.txt', sep = '\t')
ess.fillna(0, inplace=True)
ess.columns = list(map(lambda x:x.split(" ")[0], ess.columns))
Breast_Cell_line = pd.read_csv('/Users/li.13503/Desktop/heatmap/sample_info.csv')
```

```python
#extract the breast cancer cell line list
Breast_cancer_df = list(Breast_Cell_line[Breast_Cell_line['primary_disease']== 'Breast Cancer']['DepMap_ID'])
```

```python
#Subset breast cancer cell line
ess_subset = ess[ess.index.isin(Breast_cancer_df)]
#or ess_subset = ess[array]
ess_subset=ess_subset.T.sort_index(axis = 0)
```

```python
ess_subset
```

```python
# Apply the condition to each element in the DataFrame
ess_subset = ess_subset.applymap(lambda x: 1 if x > -1 else 0)
# Remove genes (rows) that are 0 in more than 95% of the samples
threshold = 0.95
num_samples = ess_subset.shape[1]
ess_subset = ess_subset[ess_subset.apply(lambda row: (row == 0).mean() <= threshold, axis=1)]
# Remove genes (rows) that are 1 in more than 95% of the samples
ess_subset = ess_subset[ess_subset.apply(lambda row: (row == 1).mean() <= threshold, axis=1)]
```

```python
Apoptosis = list(set(list(shangjia[shangjia['Pathway'] == 'Apoptosis']['Gene'])).intersection(ess_subset.index.tolist()))
Autosis = list(set(list(shangjia[shangjia['Pathway'] == 'Autosis']['Gene'])).intersection(ess_subset.index.tolist()))
Necroptosis = list(set(list(shangjia[shangjia['Pathway'] == 'Necroptosis']['Gene'])).intersection(ess_subset.index.tolist()))
Ferroptosis = list(set(list(shangjia[shangjia['Pathway'] == 'Ferroptosis']['Gene'])).intersection(ess_subset.index.tolist()))
Pyroptosis = list(set(list(shangjia[shangjia['Pathway'] == 'Pyroptosis']['Gene'])).intersection(ess_subset.index.tolist()))
Efferocytosis = list(set(list(shangjia[shangjia['Pathway'] == 'Efferocytosis']['Gene'])).intersection(ess_subset.index.tolist()))
Parthanatos = list(set(list(shangjia[shangjia['Pathway'] == 'Parthanatos']['Gene'])).intersection(ess_subset.index.tolist()))
Lysosome_CD = list(set(list(shangjia[shangjia['Pathway'] == 'Lysosome_CD']['Gene'])).intersection(ess_subset.index.tolist()))
MPT_driven_necrosis= list(set(list(shangjia[shangjia['Pathway'] == 'MPT-driven necrosis ']['Gene'])).intersection(ess_subset.index.tolist()))
Mitotic_CD = list(set(list(shangjia[shangjia['Pathway'] == 'Mitotic_CD']['Gene'])).intersection(ess_subset.index.tolist()))
Autophagy = list(set(list(shangjia[shangjia['Pathway'] == 'Autophagy\xa0']['Gene'])).intersection(ess_subset.index.tolist()))
Immunogenic = list(set(list(shangjia[shangjia['Pathway'] == '\ufeffImmunogenic']['Gene'])).intersection(ess_subset.index.tolist()))
```

```python
import seaborn
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
colors = ["#FFFF00", "#0000FF"]
custom_cmap = LinearSegmentedColormap.from_list("yellow_to_blue", colors)

seaborn.clustermap(ess_subset.loc[Apoptosis],cmap=custom_cmap, figsize=(30,12))
plt.savefig('Apoptosis.pdf', dpi=300)
seaborn.clustermap(ess_subset.loc[Autosis],cmap=custom_cmap,figsize=(30,12))
plt.savefig('Autosis.pdf', dpi=300)
seaborn.clustermap(ess_subset.loc[Necroptosis],cmap=custom_cmap,figsize=(30,12))
plt.savefig('Necroptosis.pdf', dpi=300)
seaborn.clustermap(ess_subset.loc[Ferroptosis],cmap=custom_cmap,figsize=(30,12))
plt.savefig('Ferroptosis.pdf', dpi=300)
#seaborn.clustermap(ess_subset.loc[Pyroptosis],cmap=custom_cmap,figsize=(30,12))
#plt.savefig('Pyroptosis.pdf', dpi=300)
seaborn.clustermap(ess_subset.loc[Efferocytosis],cmap=custom_cmap,figsize=(30,12))
plt.savefig('Efferocytosis.pdf', dpi=300)
#seaborn.clustermap(ess_subset.loc[Parthanatos],cmap=custom_cmap,figsize=(30,12))
#plt.savefig('Parthanatos.pdf', dpi=300)
#seaborn.clustermap(ess_subset.loc[Lysosome_CD],cmap=custom_cmap,figsize=(30,12))
#plt.savefig('Lysosome_CD.pdf', dpi=300)
#seaborn.clustermap(ess_subset.loc[MPT_driven_necrosis],cmap=custom_cmap,figsize=(30,12))
#plt.savefig('MPT_driven_necrosis.pdf', dpi=300)
seaborn.clustermap(ess_subset.loc[Mitotic_CD],cmap=custom_cmap,figsize=(30,12))
plt.savefig('Mitotic_CD.pdf', dpi=300)
seaborn.clustermap(ess_subset.loc[Autophagy],cmap=custom_cmap,figsize=(30,12))
plt.savefig('Autophagy.pdf', dpi=300)
#seaborn.clustermap(ess_subset.loc[Immunogenic],cmap=custom_cmap,figsize=(30,12))
#plt.savefig('Immunogenic.pdf', dpi=300)
```

```python
relative_freq = (ess_subset.loc[Apoptosis] == 0).mean(axis=1)
sorted_genes = relative_freq.sort_values(ascending=False)
plt.bar(sorted_genes.index, sorted_genes, color='skyblue')
plt.title('Relative Frequencies of Essentiality for Genes')
plt.xlabel('Genes (Ranked)')
plt.ylabel('Relative Frequency of Essentiality')
plt.xticks(rotation=90)
plt.grid(axis='y')
# Add threshold lines
threshold1 = 0.25
threshold2 = 0.75
plt.axhline(y=threshold1, color='black', linestyle='--', label=f'Threshold {threshold1}')
plt.axhline(y=threshold2, color='black', linestyle='--', label=f'Threshold {threshold2}')
plt.show()
```
